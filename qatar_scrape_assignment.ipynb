```python
import requests as re
from bs4 import BeautifulSoup 

url = 'https://en.wikipedia.org/wiki/Qatar'
```

```python
response = re.get(url)
response
```

```python
soup = BeautifulSoup
content = soup(response.content, 'html.parser')
content
```

```python
title = content.find('title').text
title
```

```python
parag = content.find('div', class_='mw-content-ltr mw-parser-output').text
parag
```

```python
for i in parag:
    paragraphs = i.find('p')
    print(paragraphs['p'])
```
